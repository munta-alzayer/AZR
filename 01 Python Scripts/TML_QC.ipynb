# MTDE Total Metal Loss QC Code

## Importing "input_file" and inputting the data into a dataframe

```python
#importing dependencies
import re
import pandas as pd
import numpy as np
import os

#START OF DATA characters
COMM1 = '~'
COMM2 = 'A'

# gathers the headers
# headers = []

# gathers the column names
columns = []

#boolean to stop the code from gathering everything before COMM1+COMM2
data_begin = False

#input file name
input_file = "HRDH_23\MP PROCESS.las"

#In the future you can open a file and output the result to it, or once the calculation is done by the code
# Output file
# op = open(input_file+"_output.las", "w")

# Inputting data from las file as variable td
with open(input_file, 'r') as td:
    
    #going through line by line
    for line in td:
        #Checks if the line is the beginning of the headers
        if (line[0] == COMM1) and (line[1] == COMM2): 
            
            #These lines were written when the code automatically calculated and added these curves to the code.
            #Adds these lines before the start of the headers
#             headers.append(" TNTH.inches      : 83 Total Nominal Thickness" + "\n")
#             headers.append(" TTHK.inches      : 84 Total Thickness" + "\n")
#             headers.append(" TML.%            : 85 Total Wall Loss" + "\n")
            
            #adds the column names to the headers variable
#             headers.append(line)
            
            #Split the line into distince words ex: "a b c" = "a", "b", "c"
            columns = line.split() 
            #Remove the first word = "~A" which is just a delimeter and not a name of a curve
            columns.pop(0)
            #change the boolean to allow the next if statement to enter
            data_begin = True
            continue
        
        
        #After the data begins, start inputting data into dataframe     
        if(data_begin):
            _dfs = [
                pd.DataFrame([line.split()], columns=columns, dtype=float),
                pd.read_table(td, sep='\s+', header=None, names=columns)
            ]
            df = pd.concat(_dfs, ignore_index=True)

        
#depth is set as the index of the dataframe        
df = df.set_index('Depth')
```

## Calculation Functions

### Decomissioned functions

```python
#After looking at the function use below, TOTNOM and TOTTHICK are not used below. I made this change a while back but forgot to remove these two functions
#DECOMISSIONED
def totnom(nom):
    #Total of the list nom will be saved in this variable
    totalnom = 0
    
    #loops through the list
    for x in nom:
        #any value that is the null value will be set to 0
        if x == -999.25:
            nom[nom.index(x)] = 0
        #else it will be added to the total
        else:
            totalnom += x
    #return total
    return totalnom

#Although written in a different way, totthick does the same function as totnom
#It can be consolidated to the same function in the future
def totthick(thck):
    total = 0
    for x in range(len(thck)):
        if thck[x] == -999.25:
            thck[x] = 0
        
        total += thck[x]
    return total

```

### Active Functions

```python
#Calculates the total metal loss percentage
def totml(totnom, totthick):
    #If the total nominal thickness = 0 then the result of the total metal loss is 0
    if(totnom == 0):
        return -999.25
    #calculates the result
    res = (totnom - totthick)/totnom
    #puts it into a percentage
    res *= 100
    
    #BELOW are some checks that might be reinstated. When the metal loss is 100 and when the metal loss is negative 
    #(i.e. the current thickness > thickness out of the factory)
    
    # if 100% set to -999.25
#     if(res >= 100):
#         return -999.25
    # if negative set to 0
    if(res <= 0):
        return -999.25
#     print(res)        
    
    #return result
    return res
```

## Where calculations and comparisons are made

```python
# df["NEWTNTH"] = -999.25
# df["NEWTTHK"] = -999.25
# df["NEWTML"] = -999.25

#Counters for the errors encountered
tml_errs = 0
tnth_errs = 0
tthk_errs = 0
ttl_rows = 0

#For each depth we get the rest of the data in the variable row 
for index, row in df.iterrows():
    #Add 1 to the total rows iterated through
    ttl_rows += 1
    #List of thicknesses used
    thick_used = []
    
    #CHANGE THESE INPUTS BASED ON THE CURVES IN THE INPUT FILE
    #THICK LIST AND NOM LIST HAVE TO BE THE SAME SIZES but different values pulled
    thick_list = {'4"': row.THK4, '7"': row.THK7, '9"': row.THK9,'13"': row.THK13}
    nom_list = {'4"': row.NTH4,'7"': row.NTH7, '9"': row.NTH9, '13"': row.NTH13}

    #Total thickness and Total Nominal thickness
    tot_thick = 0
    tot_nom = 0
    
    #Nominal sizes used
    nominals = [] 
    for x, y in thick_list.items():
        #Checks the list of thicknesses
        #if the value of a curve is not null and the length of the list is less than 3
        if(y != -999.25 and len(thick_used) != 3):
            #Add that thickness to the list to be displayed when an error occurs
            thick_used.append(x)
            #Add it to the total thickness
            tot_thick += y
            
            #Add nominal thickness based on the tubing added to the total thickness
            tot_nom += nom_list[x]
            #Add that size to the list of nominals used to be displayed when an error occurs
            nominals.append(nom_list[x])

    #checks if the nominal calculations are the same as the one from the file
    #If not, check if one of the values is null and set that to 0
    if(tot_nom != row.TNTH):
        nom_list = [x if x != -999.25 else 0 for x in nom_list.values()]
        tot_nom = sum(nom_list)
    
    if(tot_thick == 0): 
        tot_thick = -999.25
    #Calculate the total metal loss percentage
    tml_perc = totml(tot_nom, tot_thick)
    
#     tml_perc = round(tml_perc, 3)
#     tot_nom = round(tot_nom, 3)
#     tot_thick = round(tot_thick, 3)
    
#     df_TTHK = round(row.TTHK, 3)
#     df_TNTH = round(row.TNTH, 3)
#     df_TML = round(row.TML, 3)
    

    #All comparisons are made with the rounding to 3 decimal places
    #Comparing the tot_thickness calculated above with the one from Geoscience
    if round(tot_thick, 3) != round(row.TTHK, 3):
        #Add one if there is an error
        tthk_errs += 1
        #Print various info about the issues
        print("ISSUE WITH TOTAL THICKNESS")
        print("Depth:" , index)
        print(thick_list)
        print(thick_used)
        print("total thickness calculated = ", tot_thick)
        print("from df = ", row.TTHK)
        print("\n")

    if round(tot_nom, 3) != round(row.TNTH, 3):
        #Add one if there is an error
        tnth_errs += 1
        #Print various info about the issues
        print("ISSUE WITH NOMINAL THICKNESS")
        print("Depth:" , index)
        print("Thicknesses: ", thick_used)
        print("Nominal thicknesses:", nom_list)
        print("total nominal thickness calculated = ", tot_nom)
        print("from df = ", row.TNTH)
        print("\n")
        
    if round(tml_perc, 3) != round(row.TML, 3):
        #Add one if there is an error
        tml_errs += 1
        #Print various info about the issues
        print("ISSUE WITH TOTAL METAL LOSS CALCULATIONS")
        print("Depth:" , index)
        print("All Sizes & measurements: ", thick_list)
        print("Sizes Used for calculations:",  thick_used)
        print("Nominal Thicknesses used:" , nominals)
        print("total metal loss = ", tml_perc, "%")
        print("from df = ", row.TML, "%")
        print("\n")
        
        
print("Total # of rows: ", ttl_rows)
print("\n\nTotal issues found in the total nominal thickness calculations: ", tnth_errs)
print("\nTotal issues found in the total thickness calculations: ", tthk_errs)    
print("\nTotal issues found in the total metal loss calculations: ", tml_errs)
```
